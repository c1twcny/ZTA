```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2.1, 3, 0.01)
a = 3
b = 3
y2 = np.power(x, 3) - a*x + b

y = np.emath.sqrt(y2)
y_imag = [0 for y1 in y if y1.imag != 0.0]
y_real = [y2.real for y2 in y if y2.real != 0.0]
y_t = np.append(y_imag, y_real)
x_ext = np.append(x, x)
y_ext = np.append(y_t, -y_t)
plt.text(-2, 4, 'EC: y^2=x^3 - a*x + b')
plt.scatter(x_ext, y_ext, s=2)
plt.axhline(y=0.0, color='k')
plt.axvline(x=0.0, color='k')
plt.show()
```
